```python
import sklearn
import pandas as pd
import numpy as np
```

```python
embedding_data = pd.read_csv("../datasets/embeddings.csv")
embedding_data['merge'] = embedding_data['username'].str.lower()
users = pd.read_csv("../datasets/accounts_processed.csv", index_col=0)
users['merge'] = users['username'].str.lower()
```

```python

old_index = users[users['name'] == "Zygmunt Frankiewicz"].index[1]
users = users.drop(old_index).reset_index()
users
```

```python
users_with_embeddings = embedding_data.merge(users, on='merge')[['username_x', 'embedding', 'coalition', 'party', 'name', 'tweets_count']].rename(columns={"username_x": "username"})
```

```python
users_with_embeddings
```

```python
embeddings = np.array([np.array([np.float(i) for i in x.replace("]", "").replace("[", "").split()]) for x in users_with_embeddings['embedding'].tolist()])
```

```python
embeddings.shape
```

```python
# usernames = a['username'].tolist()
```

```python
from sklearn.manifold import TSNE
from sklearn.mixture import GaussianMixture
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler, Normalizer
```

```python
embeddings_normalized = Normalizer().fit_transform(embeddings)
embeddings_standardized = StandardScaler().fit_transform(embeddings)


tsne3d = TSNE(n_components=3).fit_transform(embeddings)
tsne2d = TSNE(n_components=2).fit_transform(embeddings)

tsne3d_standardized = TSNE(n_components=3).fit_transform(embeddings_standardized)
tsne3d_normalized = TSNE(n_components=3).fit_transform(embeddings_normalized)

tsne2d_standardized = TSNE(n_components=2).fit_transform(embeddings_standardized)
tsne2d_normalized = TSNE(n_components=2).fit_transform(embeddings_normalized)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

df = pd.DataFrame(tsne3d, columns=['x', 'y', 'z'])
df['username'] = users_with_embeddings['username']
df['coalition'] = users_with_embeddings['coalition']
fig = px.scatter_3d(df, x='x', y='y', z='z', hover_name="username", color='coalition')
fig.show()
```

```python
df = pd.DataFrame(tsne2d, columns=['x', 'y'])
df['username'] = users_with_embeddings['username']
df['coalition'] = users_with_embeddings['coalition']
fig = px.scatter(df, x='x', y='y', hover_name="username", color='coalition')
fig.show()
```

```python
from sklearn.decomposition import PCA
```

```python
pca3d = PCA(n_components=3).fit_transform(embeddings)
pca2d = PCA(n_components=2).fit_transform(embeddings)
```

```python
df = pd.DataFrame(pca3d, columns=['x', 'y', 'z'])
df['username'] = users_with_embeddings['username']
df['coalition'] = users_with_embeddings['coalition']
fig = px.scatter_3d(df, x='x', y='y', z='z', hover_name="username", color='coalition')
fig.show()
```

```python
df = pd.DataFrame(pca2d, columns=['x', 'y'])
df['username'] = users_with_embeddings['username']
df['coalition'] = users_with_embeddings['coalition']
fig = px.scatter(df, x='x', y='y', hover_name="username", color='coalition')
fig.show()
```

```python
from sklearn.preprocessing import StandardScaler, Normalizer
from sklearn.cluster import KMeans
clusters = KMeans(n_clusters=5).fit(embeddings_normalized)
x_new = clusters.transform(embeddings_normalized)

df = pd.DataFrame(tsne2d_normalized, columns=['x', 'y'])
df['username'] = users_with_embeddings['username']
df['coalition'] = users_with_embeddings['coalition']
df['cluster'] = clusters.labels_
fig = px.scatter(df, x='x', y='y', hover_name="username", color='cluster')
fig.show()
```

```python
users_with_embeddings['x_3d'] = tsne3d_normalized[:, 0]
users_with_embeddings['y_3d'] = tsne3d_normalized[:, 1]
users_with_embeddings['z_3d'] = tsne3d_normalized[:, 2]

users_with_embeddings['x_2d'] = tsne2d_normalized[:, 0]
users_with_embeddings['y_2d'] = tsne2d_normalized[:, 1]

users_with_embeddings['cluster'] = clusters.labels_

user_points = users_with_embeddings.drop(columns=['embedding'])
user_points[['username', 'x_3d', 'y_3d', 'z_3d']].rename(columns={'username': 'twitter_name', 'x_3d': 'x', 'y_3d': 'y', 'z_3d': 'z'}).to_csv('points_3d.csv', index=False)
user_points[['username', 'x_2d', 'y_2d']].rename(columns={'username': 'twitter_name', 'x_2d': 'x', 'y_2d': 'y'}).to_csv('points_2d.csv', index=False)
user_points[['username', 'name', 'party', 'coalition', 'tweets_count']].rename(columns={'username': 'twitter_name'}).to_csv('users_backend.csv', index=False)
```
